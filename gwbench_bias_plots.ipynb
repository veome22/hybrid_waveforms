```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import matplotlib


font = {'family' : 'serif',
        'size'   : 16}

matplotlib.rc('font', **font)
```

```python
df = pd.read_csv('Mc_bias_uniform_statistics.csv')
df = df.loc[~(df==0).all(axis=1)]
df
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
# cm = plt.cm.get_cmap('RdYlBu')
cm = plt.cm.get_cmap('GnBu')

sc = plt.scatter(df["M_tot"], 1/df["q"], c=df["Mc_min_faith"], cmap=cm, norm=matplotlib.colors.LogNorm())
cbar = plt.colorbar(sc)
cbar.set_label('Faithfulness')

plt.xlabel("M_tot")
plt.ylabel("1/q")
plt.title("Mc Waveform Faithfulness required")

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
# cm = plt.cm.get_cmap('RdYlBu')
cm = plt.cm.get_cmap('GnBu')

sc = plt.scatter(df["M_tot"], 1/df["q"], c=df["full_bias"]/df["Mc_stat_err"], cmap=cm, norm=matplotlib.colors.LogNorm())
cbar = plt.colorbar(sc)
cbar.set_label('bias / stat. err')

plt.xlabel("M_tot")
plt.ylabel("1/q")
plt.title("Mc Theoretical Bias \n IMRPhenomXAS vs IMRPhenomD")

plt.tight_layout()
plt.show()
```
